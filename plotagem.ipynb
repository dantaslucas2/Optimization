```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x1, x2):
    return -np.exp(-x1**2 - x2**2)
```

```python
x1 = np.linspace(-3, 3, 400)
x2 = np.linspace(-3, 3, 400)
x1, x2 = np.meshgrid(x1, x2)
z = f(x1, x2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x1, x2, z, cmap='viridis')

ax.set_title('Surface plot of $f(x_1, x_2)$')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f(x_1, x_2)$')

plt.show()
```

```python
plt.figure()
cp = plt.contourf(x1, x2, z, cmap='viridis')
plt.colorbar(cp)

plt.title('Contour plot of $f(x_1, x_2)$')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

plt.show()
```

```python
def f(x1, x2, x3, x4, x5):
    return -np.exp(-x1**2 - x2**2 - 0.5*x3**2 - 0.1*x4**4 - (x5-1)**2)
```

```python
# Fixar x3, x4, x5 e variar x1 e x2
x1 = np.linspace(-3, 3, 400)
x2 = np.linspace(-3, 3, 400)
x1, x2 = np.meshgrid(x1, x2)
x3, x4, x5 = 0, 0, 1  # Valores fixos

z = f(x1, x2, x3, x4, x5)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x1, x2, z, cmap='viridis')

ax.set_title('Surface plot of $f(x_1, x_2)$ with $x_3=0$, $x_4=0$, $x_5=1$')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f(x_1, x_2)$')

plt.show()
```

```python
plt.figure()
cp = plt.contourf(x1, x2, z, cmap='viridis')
plt.colorbar(cp)

plt.title('Contour plot of $f(x_1, x_2)$ with $x_3=0$, $x_4=0$, $x_5=1$')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir a funo f(x1, x2, x3, x4, x5)
def f(x1, x2, x3, x4, x5):
    return -np.exp(-x1**2 - x2**2 - 0.5*x3**2 - 0.1*x4**4 - (x5-1)**2)

# Intervalo de valores para as variveis
x = np.linspace(-3, 3, 100)

# Lista de variveis fixas e variveis de plotagem
variables = ['x1', 'x2', 'x3', 'x4', 'x5']
fixed_values = [0, 0, 0, 0, 1]  # Valores fixos para as variveis no plotadas

# Plotar todas as combinaes de variveis possveis
for i in range(len(variables)):
    for j in range(i + 1, len(variables)):
        # Criar grid para as duas variveis de plotagem
        xi, xj = np.meshgrid(x, x)
        
        # Copiar os valores fixos e substituir as variveis de plotagem
        fixed_values_copy = fixed_values.copy()
        xi_label = variables[i]
        xj_label = variables[j]
        
        # Mapear as variveis de plotagem para os ndices
        fixed_values_copy[i] = xi
        fixed_values_copy[j] = xj
        
        # Calcular a funo para as variveis de plotagem
        z = f(*fixed_values_copy)
        
        # Plotagem 3D
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.plot_surface(xi, xj, z, cmap='viridis')
        ax.set_title(f'Surface plot of $f({xi_label}, {xj_label})$ with fixed variables')
        ax.set_xlabel(f'${xi_label}$')
        ax.set_ylabel(f'${xj_label}$')
        ax.set_zlabel('$f$')
        plt.show()
        
        # Plotagem de contorno
        plt.figure()
        cp = plt.contourf(xi, xj, z, cmap='viridis')
        plt.colorbar(cp)
        plt.title(f'Contour plot of $f({xi_label}, {xj_label})$ with fixed variables')
        plt.xlabel(f'${xi_label}$')
        plt.ylabel(f'${xj_label}$')
        plt.show()
```
